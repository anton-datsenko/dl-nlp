```python
# Using FreqDist, find the following from text4
# words that are hapaxes (unique) and those that occur less than 5 times
# draw a plot showing mapping from word length to word frequency
# find 10 most frequent words occurring at the end of the sentence

fdist = FreqDist(text4)
print("Words that are hapaxes:", fdist.hapaxes())
print("Words that occur less than 5 times:", [i for i in fdist if fdist[i] < 5])
```

```python
plt.figure(figsize=(15, 5))
fdist.plot(100)
plt.show()
```

```python
sentences_endings = [text4[i] for i in range(len(text4)-1) if text4[i+1].endswith('.') or text4[i+1].endswith('!') or text4[i+1].endswith('?')]
print("10 most frequent words occurring at the end of the sentence:", FreqDist(sentences_endings).most_common(10))
```

```python
with open('the little prince.txt','r', encoding='utf-8') as f:
    raw = f.read()
    tokens = nltk.word_tokenize(raw)
    text = nltk.Text(tokens)
    text.concordance("other")
```

```python
text.similar('other')
```

```python
text.dispersion_plot(["little", "other", "fox", "rose", "planet"])
```

```python
fdist = FreqDist(text)
print("Words that are hapaxes:", fdist.hapaxes())
print("Words that occur less than 10 times:", [i for i in fdist if fdist[i] < 10])
```

```python
plt.figure(figsize=(15, 5))
fdist.plot(100)
plt.show()
```

```python
# Write a program that will
# collect all monetary amounts from a given text
# Example: ” first amount is $123.45, second amount is $400” −> [123.45, 400]

def collect_monetary_amounts(text):
    pattern = r'\$([0-9]+(?:\.[0-9]+)?)'
    matches = re.findall(pattern, text)
    return [float(match) if '.' in match else int(match) for match in matches]

text = "first amount is $123.45, second amount is $400, third amount is $1000.00, fourth amount is $12500"
print("Monetary amounts:", collect_monetary_amounts(text))
print("Sum of monetary amounts:", sum(collect_monetary_amounts(text)))
```

```python
json = {
    "HTTPHost" : "localhost",
    "HTTPPort" : 8545,
    "Config" : {
        "Enabled" : True,
        "Server" : "wakuv2.prod" ,
        "Nodes" : ["8.210.222.231" , "168.166.125.145"]
    }
}

# Write a program that will
# Remove entries from ”Nodes” that begin with ”8.”
# Replace each ”Server” entry of the form ServerName.Serverype with {”name”: ”ServerName”, ”type”: ”ServerType”}
# So, for example, ”Server ”: ”wakuv2.prod” will be replaced with ” Server ” : {” name ” : ” wakuv2 ” , ” type ” : ” prod ”}

def convert_json(json):
    json["Config"]["Nodes"] = [node for node in json["Config"]["Nodes"] if not node.startswith("8.")]
    json["Config"]["Server"] = { "name": json["Config"]["Server"].split('.')[0], "type": json["Config"]["Server"].split('.')[1] }
    return json

print(convert_json(json))
```
